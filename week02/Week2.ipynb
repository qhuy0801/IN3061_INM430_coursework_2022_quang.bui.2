```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the datasets
passengers = pd.read_csv("passengerData.csv", header = 0)
ticketPrices = pd.read_excel("ticketPrices.xlsx", header = 0)
```

```python
# Check to see if dataset loaded
passengers
```

```python
# Check to see if dataset loaded
ticketPrices
```

```python
# Merge the datasets
merged = pd.merge(passengers, ticketPrices, on='TicketType')

#Check to see if the dataset is loaded correctly
merged
```

```python
# Display the oldest passenger
merged.loc[merged["Age"] == merged["Age"].max()]
```

```python
# Other way to display the name of oldest passenger
# Find the maximum age
maxAge = merged['Age'].max()
# Get the names of passengers of the maximum age
oldest_names = merged['Name'][(merged.Age == maxAge)]
# Print the values
print(oldest_names.values)
```

```python
# Plot the data on a scatter plot of age vs ticket prices
plt.scatter(merged["Fare"], merged["Age"])
plt.xlabel("Fare")
plt.ylabel("Age")
plt.title("Relationship between fare and age")
plt.show()
```

```python
# Other way to plot the same plot
ax=plt.axes() # Get the axes
plt.scatter(merged.Fare, merged.Age, c = '#348abd')
ax.set_xlabel('Fare')
ax.set_ylabel('Age')
ax.set_title('Relationship between fare and age')
```

```python
# Plot only the data that shows female passengers aged 40 to 50 and who paid more than or equal to 40.
female_40 = merged.loc[(merged["Sex"] == "female") & (merged["Age"] >= 40) & (merged["Age"] <= 50) & (merged["Fare"] >= 40)]
plt.scatter(female_40["Fare"], female_40["Age"], )
plt.title("Female, aged from 40 to 50, paid equal or more than 40")
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```

```python
# Same plot but different writing method
subset = merged[((merged.Age > 40) & (merged.Age < 50) & (merged.Fare >= 40) & (merged.Sex == "female"))]

ax = plt.axes()
plt.scatter(subset.Fare, subset.Age)
ax.set_xlabel('Fare')
ax.set_ylabel('Age')
ax.set_title('Female, aged from 40 to 50, paid equal or more than 40')
```

```python
# Put mean and other descriptive statistics
stats_titanic = df_titanic.describe()
stats_titanic
```

```python
# Extract and print the mean of the age from this stats dataframe
print("The mean of the age from the stats dataframe is ",stats_titanic['Age']['mean'])

# Calculate and print the mean of the age using Series' mean() function
print("The mean of the age from Series' mean function is ",df_titanic['Age'].mean())
```

```python
# Fill with zeros
# In the case that we don't need extra dataframe as a reference, we can use
# ageWithZero = titanicDF['Age'].fillna(0, inplace=True)
ageWithZero = df_titanic['Age'].fillna(0)
fareWithZero = df_titanic['Fare'].fillna(0)

# Fill with mean
ageWithMean = df_titanic['Age'].fillna(stats_titanic['Age']['mean'])
fareWithMean = df_titanic['Fare'].fillna(df_titanic['Fare'].mean())

# This optional step will identify rows where at least one of the values is NaN with True
atLeastOneNaN = df_titanic['Age'].isnull() | df_titanic['Fare'].isnull()

# Here, I'm specifying red for imputed values and blue for the others.
# Create dictionary
colourPalette = {True: '#ff000011', False: '#0000ff11'}

# Then I'm putting the colours in the corresponding rows of a new column
colours = atLeastOneNaN.map(colourPalette)

#New I'm making a grid (1 row, 2 columns) of subplots, which share a y-axis
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

# In the first, I'm plotting the zero-imputed values
ax1.scatter(ageWithZero, fareWithZero, c=colours)
ax1.set_title('Imputed with zero')

# Then, I'm plotting the mean-imputed values
ax2.scatter(ageWithMean, fareWithMean, c=colours)
ax2.set_title('Imputed with the mean')
```

To be honest, this isn't the best data to demonstrate different imputation methods. For example, zero is clearly a bad impuation choice!

(Not surprisingly), the mean-replacements seem to work better - they seem to fitting better to the other data points. It might also be worth checking the variety in the values. If high, the mean-replacement might flatten the data, artifically reducing the variance. Also note that as the mean includes the outlier values (see point with exceptionally high values in y) it might be too high. So you might prefer to use a robust measure, such as median.

Since we will use "e_prev_100k_hi", we will replace missing values of it with the mean

```python
# Replace the missing value, using
tbDF["e_prev_100k_hi"].fillna(tbDF.e_prev_100k_hi.mean(), inplace=True)

# Plot a histogram
plt.hist(tbDF.e_prev_100k_hi, 20);
```

```python
# Select all numeric columns
tbDFNumeric = tbDF.select_dtypes(include='number')

# Get the number of columns which selected
tbDFNumeric_count = tbDFNumeric.shape[1]

# Create supplots
fig, axisRows = plt.subplots(tbDFNumeric_count//5+1, 5, figsize=(20, 20))
i = 0
for axisCols in axisRows:
    for axis in axisCols:
        if (i < tbDFNumeric_count):
            tbDFNumeric.hist(column = tbDFNumeric.columns[i], bins = 10, ax=axis)
            i = i + 1
```

```python
# Apply log on fillna column
tbDF['e_prev_100k_hi_normalised'] = tbDF.e_prev_100k_hi.apply(np.log)

# Check before / after normalisation
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.hist(tbDF.e_prev_100k_hi, 20)
ax1.set_title('Original values')
ax2.hist(tbDF.e_prev_100k_hi_normalised, 20)
ax2.set_title('Normalised values')
```

```python
# Fill all NaN
tbDFNumeric_filled = tbDFNumeric.fillna(tbDFNumeric.mean())

# Normalising using min/ max normalisation
tbDFNumeric_normalised = (tbDFNumeric_filled - tbDFNumeric_filled.min())/(tbDFNumeric_filled.max() - tbDFNumeric_filled.min())

# Convert to z-scores
tbDFNumeric_normalised_zScore = (tbDFNumeric_filled - tbDFNumeric_filled.mean())/tbDFNumeric_filled.std()
```
