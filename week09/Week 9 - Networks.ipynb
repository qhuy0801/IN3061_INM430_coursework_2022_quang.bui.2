# Analysis of Networks with NetworkX
In this exercise, we will use the NetworkX library to create our own network and then display it as a Graph. [NetworkX](https://networkx.github.io/) is a Python package for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks. It includes:

- data structures for storing graphs, digraphs, and multigraphs
- many standard graph algorithms
- network structure and analysis measures
- generators for classic graphs, random graphs, and synthetic networks
- ability for nodes to be "anything" (e.g., text, images, XML records)
- ability for edges to hold arbitrary data (e.g., weights, time-series)
See [their examples](https://networkx.github.io/documentation/stable/auto_examples/index.html) to get an idea what types of tasks can be accomplished.

- Import the library using:

- Create new empty graph:  
`G=nx.Graph()`  

Then create the graph shown above using networkX by adding nodes and edges. Use the [online documentation](https://networkx.github.io/documentation/stable/tutorial.html) for help in doing this.
- Use NetworkX's centrality measures [listed here](https://networkx.github.io/documentation/stable/reference/algorithms/centrality.html) to:
    - Find the node with highest degree centrality.
    - Find the node with highest betweenness centrality.
    - Add some weights to the edges and find the betweenness centrality once more but this time give the edge weights as parameters. Notice how things change.

```python
G=nx.Graph()

# Add nodes
G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

# Add edges
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'E'), ('C', 'D'), ('D', 'F'), ('F', 'G'), ('F', 'H')])

# Render
nx.draw(G)
```

```python
degree_c = nx.degree_centrality(G)
```
