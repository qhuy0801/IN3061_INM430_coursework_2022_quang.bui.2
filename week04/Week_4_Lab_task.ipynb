```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("censusCrimeClean.csv", header = 0)
df.head()
```

```python
medIncome = df["medIncome"]
ViolentCrimesPerPop = df["ViolentCrimesPerPop"]
```

```python
pearCorr = stats.pearsonr(medIncome, ViolentCrimesPerPop)
spearmCorr = stats.spearmanr(medIncome, ViolentCrimesPerPop)
```

```python
print(pearCorr)
print(spearmCorr)
```

```python
plt.scatter(medIncome, ViolentCrimesPerPop)
```

For the selected collumns, Person ans Spearman give 2 different value, however, the there won't be a large gap between two values (0.42 vs 0.48). However, p-values show a noticable different.
